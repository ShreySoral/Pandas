```python
#  ranking 
# The .rank() function in pandas is used to assign ranks to numeric values in a column, like scores or points. By default, it gives the average rank to tied values, which can result in decimal numbers.
df["Age"] = df["Age"].rank()                 # Default: average method
```

```python
df
```

```python
df["Rank"] = df["Age"].rank(method="dense")   # 1, 2, 2, 3
```

```python
# renaming columns and index 
df.rename(columns={"Name": "newName"}, inplace=True)
df.rename(index={0: "row1", 1: "row2"}, inplace=True)
```
